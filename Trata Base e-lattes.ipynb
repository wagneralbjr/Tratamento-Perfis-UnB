```python
import json
import pandas as pd
import numpy as np

json_normalize = pd.io.json.json_normalize
```

```python
#mostra todas as colunas
pd.set_option('display.max_columns', None)
```

```python
_json = json.load(open('unb.Perfis.Linguistica.json',encoding='UTF-8'))
```

```python
#primeiro consertamos, fazemos o ID virar campo do json e não a chave em si.

json_new = []
for key, value in _json.items():
    value['id'] = key
    json_new.append(value)
    
    
df = pd.DataFrame(json_new)
```

```python
df.head()
```

```python
#monta um dataFrame com as informações de Endereço.
def unpack_endereco(elem, df_res = None):
    tempDF = pd.DataFrame(elem.endereco_profissional,index = [0])
    return tempDF
```

```python
unpack_endereco(df.iloc[5])
```

```python
#nessa parte, monta um dataframe com as informações dos pesquisadores adicionais.
res = pd.DataFrame()
for key,value in df.iterrows():
    res = pd.concat([res,unpack_endereco(value)])
    
res = res.reset_index(drop=True)
```

```python
res.head()
```

```python
df = pd.concat([df,res.reset_index(drop=True)],axis=1)
df = df.drop('endereco_profissional',axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
# cria um df com as áreas.
_areas_df = []
for a,(key,value) in enumerate(df.iterrows()):
    _areas_df.append(pd.DataFrame(value['areas_de_atuacao'],index = [a for _ in range(0,len(value['areas_de_atuacao']))]))

area_df = pd.concat(_areas_df)
```

```python
#cria um df com as áreas com join.

joined = df.join(area_df)
```

```python
## Trata coluna producao cientifica.

_result = []
for n_id,(key,value) in enumerate(df.iterrows()):
    column = value.orientacoes_academicas
    try:
        for tipo_orientacao,list_dic in column.items():

            for elem in list_dic:
                elem['tipo_orientacao'] = tipo_orientacao
                aux_df = pd.DataFrame(elem, index=[n_id for _ in range(0 , len(elem) )])
                #print(aux_df)
                _result.append(aux_df)
    except:
        continue
        
prod_cientifica = pd.concat(_result)
```

```python
#producao_bibiografica
_result = []
for n_id,(key,value) in enumerate(df.iterrows()):
    column = value.producao_bibiografica
    try:
        for tipo_orientacao,list_dic in column.items():
            for elem in list_dic:
                elem['tipo_documento'] = tipo_orientacao
                aux_df = pd.DataFrame(elem, index=[n_id for _ in range(0 , len(elem) )])
                #print(aux_df)
                _result.append(aux_df)
    except Exception as e:
        print(e)
        break
        continue
        
prod_bibiografica = pd.concat(_result)
```
