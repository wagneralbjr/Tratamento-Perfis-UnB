```python
#primeiro consertamos, fazemos o ID virar campo do json e não a chave em si.

json_new = []
for key, value in _json.items():
    value['id'] = key
    json_new.append(value)
    
    
df = pd.DataFrame(json_new)
```

```python
#monta um dataFrame com as informações de Endereço.
def unpack_endereco(elem, df_res = None):
    tempDF = pd.DataFrame(elem.endereco_profissional,index = [0])
    return tempDF
```

```python
unpack_endereco(df.iloc[5])
```

```python
#nessa parte, monta um dataframe com as informações dos pesquisadores adicionais.
res = pd.DataFrame()
for key,value in df.iterrows():
    res = pd.concat([res,unpack_endereco(value)])
    
res = res.reset_index(drop=True)
```

```python
res.head()
```

```python
df = pd.concat([df,res.reset_index(drop=True)],axis=1)
df = df.drop('endereco_profissional',axis=1)
```
