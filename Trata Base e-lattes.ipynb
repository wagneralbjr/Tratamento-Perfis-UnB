```python
_json = json.load(open('unb.Perfis.Linguistica.json',encoding='UTF-8'))
```

```python
#primeiro consertamos, fazemos o ID virar campo do json e não a chave em si.

json_new = []
for key, value in _json.items():
    value['id'] = key
    json_new.append(value)
    
    
df = pd.DataFrame(json_new)
```

```python
df.head()
```

```python
#monta um dataFrame com as informações de Endereço.
def unpack_endereco(elem, df_res = None):
    tempDF = pd.DataFrame(elem.endereco_profissional,index = [0])
    return tempDF


#nessa parte, monta um dataframe com as informações dos pesquisadores adicionais.
res = pd.DataFrame()
for key,value in df.iterrows():
    res = pd.concat([res,unpack_endereco(value)])
    
res = res.reset_index(drop=True)
```

```python
# df com informações dos endereços.
res.head()
```

```python
#concatenando o data frame de endereços com o DF geral.
df = pd.concat([df,res.reset_index(drop=True)],axis=1)
df = df.drop('endereco_profissional',axis=1)
```

```python
# cria um df com as áreas.
_areas_df = []
for a,(key,value) in enumerate(df.iterrows()):
    _areas_df.append(pd.DataFrame(value['areas_de_atuacao'],index = [a for _ in range(0,len(value['areas_de_atuacao']))]))

area_df = pd.concat(_areas_df)
```

```python
'''Exemplo de como combinar o data frame de áreas com o df Principal,
fazemos um Join entre eles utilizando como chave o index.'''


joined = df.join(area_df)
```

```python
def _salva_dict(dic, entry, kind):
    """ Salva no dicionário separando por tipo."""
    if kind in dic:
        dic[kind].append(entry)
    else:
        dic[kind] = list(entry)

    return

def _trata_dict(dic):
    """ Elimina os tipos q não viraram DataFrame"""
    for key, value in dic.items():
        res = []
        for elem in value:
            if isinstance(elem, pd.DataFrame):
                res.append(elem)
        dic[key] = res
        
    return
       
""" Trata a coluna orientações academicas..
    Não é possível tratar todos os elementos de uma vez por campos diferentes.
"""
_result = {}
for n_id,(key,value) in enumerate(df.iterrows()):
    column = value.orientacoes_academicas # coluna a ser retirada.

    try:
        for tipo_orientacao,list_dic in column.items():
            for elem in list_dic:
                elem['tipo_orientacao'] = tipo_orientacao
                aux_df = pd.DataFrame(elem)
                aux_df.index = [n_id for _ in range ( len(aux_df.index) )]                
                _salva_dict(_result, aux_df, tipo_orientacao)
    except Exception as e:
        print(elem)
        break

_trata_dict(_result) #remove os elementos que não formam data-frames.

""" Para retirar os dfs de orientação, acesse-os com as seguintes keys.
dentro de _result.


dict_keys(['ORIENTACAO_CONCLUIDA_MESTRADO', 'ORIENTACAO_EM_ANDAMENTO_DOUTORADO', 
'ORIENTACAO_EM_ANDAMENTO_MESTRADO', 'OUTRAS_ORIENTACOES_CONCLUIDAS',
'ORIENTACAO_EM_ANDAMENTO_GRADUACAO', 'ORIENTACAO_EM_ANDAMENTO_INICIACAO_CIENTIFICA',
'ORIENTACAO_CONCLUIDA_DOUTORADO', 'ORIENTACAO_CONCLUIDA_POS_DOUTORADO'])

"""

# df_ori_concl_mestrado = pd.concat(_prod_cientifica['ORIENTACAO_CONCLUIDA_MESTRADO'])
# df_ori_andamento_doutorado = pd.concat(_prod_cientifica['ORIENTACAO_EM_ANDAMENTO_DOUTORADO'])
# df_ori_andamento_graduacao = pd.concat(_prod_cientifica['ORIENTACAO_EM_ANDAMENTO_GRADUACAO'])
# df_outras_orientacoes_concluidas = pd.concat(_prod_cientifica['OUTRAS_ORIENTACOES_CONCLUIDAS'])

""" Para juntar as orientações academicas junto ao orientador, utilize um join como abaixo.
    #dataframe principal join df_prod_cientifica.
    df.join(df_prod_cientifica)
"""

lis = []
for key,value in _result.items():
    for elem in _result[key]:
        lis.append(elem)

df_orientacoes_academicas = pd.concat(lis)  


```
